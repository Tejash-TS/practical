```python
# Assignment 9: Linear Regression using Deep Neural Network
# Implement Boston Housing Price Prediction using Deep Neural Network

import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import boston_housing

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python

# Load IMDB Dataset Locally
# imdb.npz should be present in project folder

data = np.load("imdb.npz", allow_pickle=True)

X_train = data['x_train']
y_train = data['y_train']

X_test = data['x_test']
y_test = data['y_test']

print("Training Samples:", len(X_train))
print("Testing Samples :", len(X_test))
```

```python
# ---------------------------------------------------
# Build Deep Neural Network Model
# ---------------------------------------------------

model = Sequential()

# Input + Hidden Layer 1
model.add(Dense(64, activation='relu',
                input_shape=(X_train.shape[1],)))

# Hidden Layer 2
model.add(Dense(32, activation='relu'))

# Hidden Layer 3
model.add(Dense(16, activation='relu'))

# Output Layer
# Single neuron because this is regression
model.add(Dense(1))
```

```python
# ---------------------------------------------------
# Compile Model
# ---------------------------------------------------

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
# ---------------------------------------------------
# Train Model
# ---------------------------------------------------

history = model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

```python
# ---------------------------------------------------
# Predict on Test Data
# ---------------------------------------------------

predictions = model.predict(X_test)

# ---------------------------------------------------
# Calculate Mean Squared Error
# ---------------------------------------------------

mse = mean_squared_error(y_test, predictions)

print("\nMean Squared Error:", mse)
```

```python
# ---------------------------------------------------
# Display Actual vs Predicted Prices
# ---------------------------------------------------

print("\nActual Price vs Predicted Price\n")

for i in range(10):
    print(f"Actual: {y_test[i]:.2f} \t Predicted: {predictions[i][0]:.2f}")
```
